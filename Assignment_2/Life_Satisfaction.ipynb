### Q1 (5 marks)
Load the dataset `lifesat.csv` and display the first 5 rows.

### Q2 (5 marks)
Print the dataset's basic information and summary statistics.

```python
print("--- Info ---")
df.info() # Print concise summary / Вывести краткую сводку

print("\n--- Summary Statistics ---")
df.describe() # Generate descriptive statistics / Генерация описательной статистики
```

### Q3 (10 marks)
Plot GDP per capita vs. Life Satisfaction. Add appropriate labels and discuss the observed relationship.

```python
plt.figure(figsize=(10, 6)) # Set figure size / Задать размер фигуры
sns.scatterplot(x='GDP per capita', y='Life satisfaction', data=df, s=100, color='teal') # Scatter plot / Точечная диаграмма
plt.title('GDP per Capita vs. Life Satisfaction', fontsize=16) # Title / Заголовок
plt.xlabel('GDP per Capita (USD)', fontsize=12) # X-axis label / Подпись оси X
plt.ylabel('Life Satisfaction', fontsize=12) # Y-axis label / Подпись оси Y
plt.grid(True, linestyle='--', alpha=0.7) # Add grid / Добавить сетку
plt.show() # Display plot / Показать график
```

## Part 2: Linear Regression Model (30 marks)

### Q4 (5 marks)
Extract the feature matrix `X` and target vector `y`. Print their shapes.

```python
X = df[['GDP per capita']].values # Feature matrix (2D array) / Матрица признаков (2D массив)
y = df['Life satisfaction'].values # Target vector (1D array) / Вектор целевой переменной (1D массив)

print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
```

### Q5 (10 marks)
Train a Linear Regression model. Display the model's coefficient and intercept.

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression() # Initialize model / Инициализация модели
lin_reg.fit(X, y) # Train model / Обучить модель

print(f"Intercept (theta_0): {lin_reg.intercept_:.4f}") # Bias term / Сдвиг
print(f"Coefficient (theta_1): {lin_reg.coef_[0]:.8f}") # Slope / Наклон
```

### Q6 (10 marks)
Plot the regression line over the scatter plot of the data.

```python
plt.figure(figsize=(10, 6)) # Set figure size / Задать размер фигуры
sns.scatterplot(x='GDP per capita', y='Life satisfaction', data=df, s=100, color='teal', label='Data') # Scatter plot / Точечная диаграмма

# Plot regression line
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_pred_line = lin_reg.predict(X_range) # Predict / Предсказать
plt.plot(X_range, y_pred_line, color='red', linewidth=2, label='Linear Regression')

plt.title('Linear Regression: GDP vs Life Satisfaction', fontsize=16) # Title / Заголовок
plt.xlabel('GDP per Capita (USD)') # X-axis label / Подпись оси X
plt.ylabel('Life Satisfaction') # Y-axis label / Подпись оси Y
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7) # Add grid / Добавить сетку
plt.show() # Display plot / Показать график
```

### Q7 (5 marks)
Predict Life Satisfaction for a GDP of $37,655.2. Comment on the result.

```python
X_new = [[37655.2]]
pred_lin = lin_reg.predict(X_new)[0] # Predict / Предсказать

print(f"Predicted Life Satisfaction for GDP $37,655.2: {pred_lin:.2f}")
```

**Comment:**
The model predicts a life satisfaction score based on the linear trend. This value represents the "average" expected satisfaction for a country with that specific GDP, assuming the linear relationship holds true.

## Part 3: K-Nearest Neighbors Regression (25 marks)

### Q8 (5 marks)
Train a `KNeighborsRegressor` with `n_neighbors=3`.

```python
from sklearn.neighbors import KNeighborsRegressor

knn_reg = KNeighborsRegressor(n_neighbors=3) # Initialize KNN / Инициализация KNN
knn_reg.fit(X, y) # Train KNN / Обучить KNN
```

### Q9 (10 marks)
Predict Life Satisfaction for a GDP of $37,655.2 using the KNN model and compare it with the Linear Regression prediction.

```python
pred_knn = knn_reg.predict(X_new)[0]

print(f"Linear Regression Prediction: {pred_lin:.2f}")
print(f"KNN (k=3) Prediction:       {pred_knn:.2f}")
print(f"Difference:                 {abs(pred_lin - pred_knn):.2f}")
```

### Q10 (10 marks)
Evaluate the model with `n_neighbors` values of [1, 3, 5, 10]. Plot the results to compare performance.

```python
k_values = [1, 3, 5, 10]
predictions = []

plt.figure(figsize=(12, 8)) # Set figure size / Задать размер фигуры
sns.scatterplot(x='GDP per capita', y='Life satisfaction', data=df, s=100, color='teal', label='Data', alpha=0.6) # Scatter plot / Точечная диаграмма

X_range = np.linspace(X.min(), X.max(), 500).reshape(-1, 1)

colors = ['orange', 'green', 'purple', 'brown']

for k, color in zip(k_values, colors):
    model = KNeighborsRegressor(n_neighbors=k)
    model.fit(X, y)
    y_pred = model.predict(X_range)
    plt.plot(X_range, y_pred, label=f'k={k}', linewidth=2, color=color)
    
    # Predict for our specific point
    pred = model.predict(X_new)[0]
    predictions.append(pred)
    print(f"Prediction for k={k}: {pred:.2f}")

plt.title('KNN Regression with Different k Values', fontsize=16) # Title / Заголовок
plt.xlabel('GDP per Capita') # X-axis label / Подпись оси X
plt.ylabel('Life Satisfaction') # Y-axis label / Подпись оси Y
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7) # Add grid / Добавить сетку
plt.show() # Display plot / Показать график
```
