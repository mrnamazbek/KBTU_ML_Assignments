## Task 1: Create Normal Headings (5 marks)
Write Markdown to create:
- A level 1 heading: "My Title"
- A level 2 heading: "My Subtitle"
- A level 3 heading: "My Section"

## Task 2: Center a Heading (10 marks)
Create a centered heading: "Welcome to My Notebook".
